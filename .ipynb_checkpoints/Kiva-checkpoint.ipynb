```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
kiva = pd.read_csv("Kiva Loan.csv")
```

```python
kiva.head(2)
```

```python
kiva.info()
```

```python
kiva.describe()
```

```python
kiva.describe(include=['O'])
```

```python
#checking the missing data.
missing_data = kiva.isnull().sum().sort_values(ascending = True)
percent = (kiva.isnull().sum()/kiva.isnull().count()).sort_values(ascending = False)
missing_values = pd.concat([missing_data, percent],axis=1,keys=['Total Missing values', 'Percentage'])
missing_values
```

```python
#sector-wise analysis
sector_count = kiva['Sector'].value_counts()
sns.barplot(y=sector_count.index, x = sector_count.values)
for i, j in enumerate(sector_count.values):
    plt.text(1, i, j, color='k', fontsize=12)
plt.xlabel("Number of Loans")
plt.ylabel("Sectors")
plt.title("Number of Loans made in different Sectors")
```

```python
#Status of the loan
status_count = kiva['Status'].value_counts()
sns.barplot(status_count.values, status_count.index)
for i, j in enumerate(status_count.values):
    plt.text(5000, i, j, color='k', fontsize=12)
plt.xlabel("Count of the Status")
plt.ylabel("Loan Status")
plt.title("Loan Status of differnt Loans")
```

```python
kiva['Status'].value_counts().plot(kind='pie', figsize=(10,10)) #Not very informative
```

```python
#Count of Activity for which Loan was asked
activity_count = kiva['Activity'].value_counts().head(12)
sns.barplot(activity_count.values, activity_count.index)
for i, j in enumerate(activity_count.values):
    plt.text(1, i, j, color='k', fontsize=12)
plt.xlabel("Count of the activity")
plt.ylabel("Activity")
plt.title("Count of Top 12 Activity");
```

```python
#Count of loan recieved by each country
country_count = kiva['Country'].value_counts().head(12)
sns.barplot(country_count.values, country_count.index)
for i, j in enumerate(country_count.values):
    plt.text(1, i, j, color='k', fontsize=12)
plt.xlabel("Count of the Loan")
plt.ylabel("Country")
plt.title("Top 12 Country to recieve Loan")
```

```python
#Distribution of Funded Amount
plt.figure(figsize = (12, 8))
sns.distplot(kiva['Funded Amount'])
plt.xlim(0,5000)
```

```python
plt.figure(figsize = (12,8))
plt.scatter(range(kiva.shape[0]), np.sort(kiva['Funded Amount'].values))
plt.xlabel("Count")
plt.ylabel("Funded Amount")
plt.title("Distribution of Funded Amount");
```

```python
#Distribution of Loan Amount
sns.distplot(kiva['Loan Amount'])
```

```python
#Distribution of Loan Amount
plt.scatter(range(kiva.shape[0]), np.sort(kiva['Loan Amount'].values))
plt.xlabel("Count")
plt.ylabel("Loan Amount")
plt.title("Distribution of Loan Amount");
```

```python
#Distribution of Sectors
import squarify
plt.figure(figsize=(15,12))
squarify.plot(sector_count.values, label=sector_count.index, value=sector_count.values)
plt.title("Distribution of Sectors");
```

```python
#Use of the Loans
use_count = kiva['Use'].value_counts().head(12)
sns.barplot(use_count.values, use_count.index)
for i, j in enumerate(use_count.values):
    plt.text(1, i, j, color='k', fontsize=12)
plt.xlabel("Count of the Loans")
plt.ylabel("Different Uses of Loan");
```
